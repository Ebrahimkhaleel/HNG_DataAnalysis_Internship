```python
import pandas as pd
from dateutil import parser

# Load the datasets
flood_history = pd.read_csv('flood history in lagos.csv')
weather_data = pd.read_csv('lagos_weeather_data_2002_to_2024.csv')

# Display the first few rows of each dataset
print(flood_history.head())
print(weather_data.head())
```

```python
# Inspect the flood history dataset
print(flood_history.info())
print(flood_history.describe())

# Inspect the weather data dataset
print(weather_data.info())
print(weather_data.describe())

# Check for missing values
print(flood_history.isnull().sum())
print(weather_data.isnull().sum())
```

```python
# Drop unnecessary columns
flood_history_cleaned = flood_history.drop(columns=['Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15'])

# Drop rows with missing values in crucial columns
flood_history_cleaned = flood_history_cleaned.dropna(subset=['Duration(days)', 'Moon phase', 'Wind speed', 'Wind direction'])

# # Convert Date column in flood history to datetime
flood_history_cleaned['Date'] = pd.to_datetime(flood_history_cleaned['Date'], format='%d-%b-%y')

# # Fill missing values in weather data with column mean
#weather_data_cleaned = weather_data.fillna(weather_data.mean())

# # Convert datetime column in weather data to datetime format with the correct format
def parse_date(date_str):
    return parser.parse(date_str, dayfirst=True)

# Apply the parse_date function to each date
weather_data['Datetime'] = weather_data['datetime'].apply(parse_date)
weather_data

# # Rename the datetime column to Date for merging
weather_data = weather_data.rename(columns={'Datetime': 'Date'})

# # # Merge datasets on the Date column
merged_data = pd.merge(weather_data, flood_history_cleaned, on='Date', how='left')

# # # Save the merged dataset
merged_data.to_csv('merged1.csv', index=False)

# # # Display the merged dataset
# print(merged_data.info())
# print(merged_data.head())

```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Assuming 'Moon phase' is a categorical column, encode it if it exists


# Normalize/scale the features

X = merged_data.drop(columns=['Date', 'Flood','datetime'])


# Target variable
target = merged_data['Flood']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.2, random_state=42)

# Model selection
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Predict on test set
y_pred = model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))


prediction = model.predict()

print("Flood Prediction:", prediction)
```
